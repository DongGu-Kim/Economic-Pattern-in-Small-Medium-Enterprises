## State Level of Data Analysis

state_merge_data1.csv dataset covers the year of 2016.It includes total numbers of graduates in Engineering and science major,total number of employed person,us population and sales tax by state.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
#using matplotlibâ€™s ggplot style
plt.style.use('ggplot')
import seaborn as sns
sns.set(color_codes=True)
```

```python
path ='/Users/gyang/Desktop/yang_project/capstone/data/'
```

```python
# Load dataset 
state_merge_data1 = pd.read_csv(path + 'state_merge_data1.csv')
state_merge_data2 = pd.read_csv(path + 'state_merge_data2.csv')
#state_merge_data.set_index("state", inplace = True)
```

```python
state_merge_data =state_merge_data1.merge(state_merge_data2,on='state')
```

```python
sns.set()

state_merge_data.sort_values(['us_population'], ascending=False, inplace=True)

f, ax = plt.subplots(nrows=1, ncols=2,figsize=(14, 15))

#add col 1 -- us population
sns.set_color_codes("muted")
sns.barplot(x='us_population', y='state', data =state_merge_data,color="b",ax=ax[0])
ax[0].set(ylabel="",xlabel="US Population")
sns.despine(left=True, bottom=True)

#add col 2 -- total number of employment 
state_merge_data.sort_values(['total_employed'], ascending=False, inplace=True)
sns.set_color_codes("muted")
sns.barplot(x='total_employed', y='state', data=state_merge_data,color="b",ax=ax[1])
ax[1].set(ylabel="",xlabel="Total Number of Employment")
sns.despine(left=True, bottom=True)
```

```python
sns.set()
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(8, 15))

state_merge_data.sort_values(['us_population'], ascending=False, inplace=True)

sns.set_color_codes("pastel")
sns.barplot(x="total_employed", y="state", data=state_merge_data,
            label="Total Employment", color="b")

sns.set_color_codes("muted")
sns.barplot(x="totals_science", y="state", data=state_merge_data,
            label="Total Science Grads", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="",
       xlabel="Total Employment (Total Science Grads)")
sns.despine(left=True, bottom=True)
```
